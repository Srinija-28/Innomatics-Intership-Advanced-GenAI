```python
#1.USER LOGIN CHECK
username="admin"
password="1234"
if username=="admin" and password=="1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2.PASS/FAIL ANALYZER
marks=[45,78,90,33,60]
pass_count=0
fail_count=0
for score in marks:
    if score>=50:
        pass_count+=1
    else:
        fail_count+=1
print("total no.of students:",len(marks))
print("pass count of students:",pass_count)
print("fail count of students:",fail_count)
```

```python
#3.SIMPLE DATA CLEANER
names=["Alice","bob","CHARLIE"]
#strip()-->removes whitespace
#lower()-->converts to lower case
cleaned_list=[name.strip().lower() for name in names]
print(cleaned_list)
```

```python
#4.Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length=len(msg)
    print(msg,length)
    if length>10:
        print("this message is longer than 10 characters")
```
