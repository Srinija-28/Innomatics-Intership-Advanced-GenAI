```python
#1.Unique Words in a Sentence
sentence = "python is easy and python is powerful"
#splitting sentence
words = sentence.split()
#extracting unique words
unique_words_set = set(words)
print("unique words:",unique_words_set)
#count of unique words
print("count of unique words:",len(unique_words_set))
```

```python
#Problem 2: Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
max_salary=0
highest_employee=""
for name,salary in employees.items():
    if salary > max_salary:
        max_salary=salary
        highest_employee=name
print("Highest Salary:",highest_employee,max_salary)
```

```python
#Problem 3: Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
max=numbers[0]
min=numbers[0]
for num in numbers:
    if num>max:
        max=number
    if num<min:
        min=number
print("List:",numbers)
print("Max:",max)
print("Min:",min)
```

```python
#Problem 4: Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count=0
for price in prices:
#product price above 1000
    if price>1000:
        count += 1
print("Products above 1000:",count)        
```

```python
#Problem 5: Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
#count of present
present_count=attendance.count("P")
total_days=len(attendance)
#calculating percentage
percentage=(present_count/total_days)*100
print("Attendence Percentage:",percentage)
```

```python
#Problem 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
#convert list into set
phn_number_set=set(phone_numbers)
#converting back to list to print
unique_phn_numbers=list(phn_number_set)
print("Unique phone numbers:",unique_phn_numbers)
```

```python
#Problem 7: Count Character Frequency
def count_char_frequency(input):
    dict={}
    for char in input:
        if char in dict:
            dict[char] += 1
        else:
            dict[char] = 1
    return dict       
input='ppython'
output=count_char_frequency(input)
print(output)
```
