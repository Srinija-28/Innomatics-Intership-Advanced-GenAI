```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans=[199,299,399,599]
    while True:
        amount=int(input("Enter recharge amount:"))
        if amount<50:
            print("Invalid:Amount must be 50rs or more.")
        elif amount not in valid_plans:
            print("Invalid:Not a valid plan.")
        else:
            print("Recharge Successful")
            break
validate_recharge()            
```

```python
#Problem 2: Inventory Reorder Alert System
def checking(products):
    for product,stock in products.items():
        if stock<15:
            print(f"Reorder Alert for {product}")
        else:
            print(f"Stock OK for {products}")
data={"Laptops":25,"Phones":12,"Tablets":50,"Monitors":8}
checking(data)
```

```python
#Problem 3: Student Result Processing System
def marks(marks_list):
    total_marks=0
    for mark in marks_list:
        total_marks+=mark
    average_marks=total_marks/len(marks_list)
    if average_marks>=50:
        result="Pass"
    else:
        result="Fail"
    return result
#example
student_marks=[45,67,34,89,50]
final_result=marks(student_marks)
print(f"The average mark is {sum(student_marks)/len(student_marks)},The result is:**{final_result}**")
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(km,is_peak):
    fare=50+(km*12)#Base fare +rate per km
    if is_peak=='yes':
        fare=fare*1.25  #add 25% extra for peak hour
    return fare
while True:
    #Get user inputs
    dist=float(input("Enter distance in km:"))
    peak=input("Is it peak hour?(yes/no):").lower()
    #calculate and show result
    total=calculate_fare(dist,peak)
    print(f"Total Fare:{total}rs")
    #retry option
    retry=input("Calculate again?(yes/no):").lower()
    if retry!="yes":
        break
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def check_attendance(attendance_list):
    present_count=0
    #use loop to count present days
    for day in attendance_list:
        if day=='P':
            present_count+=1
    #calculate attendence percentage
    total_days=len(attendance_list)
    percentage=(present_count/total_days)*100
    #use return statement
    if percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"
#example
list=['P','P','P','A']#75% present
print(check_attendance(list))
```

```python
#Problem 6: Password Strength Checker
def check_password_strength(password):
    #rule-1:min len=8
    if len(password)<8:
        return False
    #rule 2:must contain at least one digit
    has_digit=False
    for char in password:
        if char.isdigit():
            has_digit=True
            break
    if not has_digit:
        return False
    #rule-3:must contain at least one special char(@#$)
    special_chars="@#$"
    has_special=False
    for char in password:
        if char in special_chars:
            has_special=True
            break
    if not has_special:
       return False
   #all rules oassed
    return True
#example usage:
print(check_password_strength("MyPassword@"))
    
```
