```python
#Problem Statement 1:Smart Parking Lot Management System
def manage_parking(capacity,logs):
    currently_parked=0
    #process each entry in the logs
    for action in logs:
        if action == "IN":
            currently_parked += 1
        elif action == "OUT":
            currently_parked=max(0,currently_parked-1)
   #Determine parking status 
    if currently_parked >= capacity:
        status="Full/Alert:Capacity Exceeded"
    else:
        status="Available"
    return currently_parked,status
#Input
parking_capacity=50
vehicle_logs=["IN","IN","IN","OUT","IN","IN","OUT"]
parked_count,parking_status=manage_parking(parking_capacity,vehicle_logs)
#output
print(f"Currently Parked Vehicles:{parked_count}")
print(f"Parking Status:{parking_status}")
```

```python
#Problem Statement 2:Online Food Delivery Time Estimator
def delivery_time(dist_km,weather_cond,traffic_level):
    #assuming base time with an avg speed
    base_time_per_km=5
    estimated_time=dist_km*base_time_per_km
    #traffic delays
    if traffic_level.lower()=="high":
        estimated_time+=10
    elif traffic_level.lower()=="medium":
        estimated_time+=5
    #weather delays
    if weather_cond.lower()=="rainy":
        estimated_time+=5
    elif weather_cond.lower()=="stormy":
        estimated_time+=15
    return estimated_time
#input
distance=11
traffic="High"
weather="Rainy"
#calculation & output
final_eta=delivery_time(distance,traffic,weather)
print(f"Estimated Delivery Time:{final_eta}minutes")
    
```

```python
#Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
def analyze_occupancy(total_seats,booked_seats_list):
    #1.calculate occupancy percentage
    booked_count=len(booked_seats_list)
    occupancy_percent=(booked_count/total_seats)*100
    #2.determine show status
    if occupancy_percent==100:
        status="Housefull"
    elif occupancy_percent>=75:
        status="Almost Full"
    else:
        status="Seats Available"
#3.Suggest opening additional shows
    suggestion="Recommended" if occupancy_percent>80 else "Not needed yet"
    return occupancy_percent,status,suggestion
#sammple input
total_seats=200
#creating list with 150 entries
booked_seats=[1]*150
#get result
percent,status,suggestion=analyze_occupancy(total_seats,booked_seats)
#expected output
print(f"Occupancy:{percent:.0f}%")
print(f"Show Status:{status}")
print(f"Additional Shows:{suggestion}")
```

```python
#Problem Statement 4:Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    #calculate average CPU load
    if not cpu_readings:
        return "No data provided"
    avg_load=sum(cpu_readings)/len(cpu_readings)
    #classifying based on rules
    if avg_load<50:
        status="Normal"
    elif 50<=avg_load<=80:
        status="Warning"
    else:
        status="Critical"
    #result
    print(f"Average CPU load:{avg_load:.0f}%")
    print(f"Server Status:{status}")
#input
cpu_data=[45,60,70,85,90]
#fun call
classify_server_load(cpu_data)
    
```

```python
#Problem Statement 5:Smart Classroom Resource Usage Monitor
def monitor_resources(usage_dict):
    #set a threshold for overuse
    threshold=8
    overused=[]
    energy_alert="No"
    for resource,hours in usage_dict.items():
        if hours>threshold:
            overused.append(resource)
            energy_alert="Yes"
    #Formating output
    print(f"Overused Resources:{', '.join(overused)}")
    print(f"Energy Alert:{energy_alert}")
#input    
resource_usage={
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}
#calling fun
monitor_resources(resource_usage)
```

```python
#Problem Statement 6: Online Event Registration Capacity Controller
def manage_registrations(capacity,registrations):
    #determine confirmed and waitlisted users
    if registrations > capacity:
        confirmed=capacity
        waitlisted=registrations-capacity
        status="Closed"
    else:
        confirmed=registrations
        waitlisted=5
        status="Open"
   #printing output
    print(f"Confirmed Registrations:{confirmed}")
    print(f"Waitlisted Users:{waitlisted}")
    print(f"Registration Status:{status}")
#input
event_capacity=100
total_registrations=100
#calling function
manage_registrations(event_capacity,total_registrations)

    
```
