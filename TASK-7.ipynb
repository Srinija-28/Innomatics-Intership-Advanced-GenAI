```python
#Problem 1: Smart Light Controller
class Smartlight:
    def __init__(self,name,status):
        self.name=name
        self.status=status
    def result(self):
       print(f"{self.name} is {self.status}")
#creating obj
light_name="Bedroom Light"
action="ON"
#creating object
my_light=Smartlight(light_name,action)
#method calling
my_light.result()
```

```python
#Problem 2: Employee ID Card System
class Employee:
    def __init__(self,name,emp_id,dept):
        self.name=name
        self.emp_id=emp_id
        self.dept=dept
    def result(self):
        print("Employee ID Card")
        print("Name",self.name)
        print("Emp_id",self.emp_id)
        print("Department",self.dept)
#input
emp_name="Rahul"
id="EMP102"
department="IT"
#creating an object
emp1=Employee(emp_name,id,department)
#method calling 
emp1.result()
```

```python
#Problem 3: Mobile Contact Record
class Contact:
    def __init__(self,name,phn_number):
        self.name=name
        self.phn_number=phn_number
    def result(self):
       print("Contact Saved")
       print("Name:",self.name)
       print("Phone:",self.phn_number)
#creating object
contact1=Contact("Anita","9876543210")
#method calling
contact1.result()
```

```python
#Problem 4: Product Price Tag Generator
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def result(self):
        print("Product:",self.name)
        print("Price:",self.price)
#input        
product_name="Headphones"
product_price=2499
#creating object
tag=Product(product_name,product_price)
#method calling
tag.result()
```

```python
#Problem 5: Movie Rating Display System
class Movie:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def result(self):
        print("Movie:",self.name)
        print("Rating:",self.rating)
#input
movie_name="Inception"
movie_rating="4.8/5"
#creating object
movie_obj=Movie(movie_name,movie_rating)
#method calling
movie_obj.result()
        
```

```python
#Problem 6: Delivery Address Manager
class DeliveryAddressManager:
    def __init__(self,name,address):
        self.name=name
        self.address=address
    def result(self):
        print("Delivery Details:")
        print("Customer:",self.name)
        print("Address:",self.address)
#input
Customer_Name="Suman"
Address="Hyderabad"
#creating object
delivery=DeliveryAddressManager(Customer_Name,Address)
#method calling
delivery.result()


```
